## 4. FEATURE IMPORTANCE

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder

# Encoding categorical variables
label_encoder = LabelEncoder()   #categorical data converted into numerical data
df['Product ID'] = label_encoder.fit_transform(df['Product ID'])
df['Type'] = label_encoder.fit_transform(df['Type'])
df['Failure Type'] = label_encoder.fit_transform(df['Failure Type'])

# Define features and target
features = df.drop('Target', axis=1)
target = df['Target']

# Train the model
model = RandomForestClassifier()
model.fit(features, target)

# Plot feature importances
feature_importances = pd.Series(model.feature_importances_, index=features.columns)
feature_importances.nlargest(10).plot(kind='barh', color='gray')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.title('Top 10 Feature Importances')
plt.show()
```

‚ö™The visual shows which features are most influential in making predictions.
‚ö™This helps in understanding how the model makes decisions.

## SMOTE TECHNIQUE

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(sampling_strategy="minority")
predictors,label = smote.fit_resample(predictors, label)
label.value_counts()
```

## Why Use SMOTE?

SMOTE helps in addressing the issue of class imbalance in your dataset by generating synthetic examples for the minority class. This can lead to better performance of machine learning models, as they can learn more effectively from a balanced dataset.

Here we created a function to check which Machine Learning Algorithm will be the best fit for our model. We will select the one with highest accuracy.

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(predictors,label,test_size=0.3,random_state=42)

from sklearn.linear_model import LogisticRegression                #LogisticRegression from sklearn.linear_model.
logreg = LogisticRegression()

from sklearn.neighbors import KNeighborsClassifier                 #KNeighborsClassifier from sklearn.neighbors
knn = KNeighborsClassifier()

from sklearn.tree import DecisionTreeClassifier                    #DecisionTreeClassifier from sklearn.tree
dt = DecisionTreeClassifier()

from sklearn.metrics import accuracy_score,classification_report   #This line imports the accuracy_score and classification_report functions
```

```python
from sklearn.metrics import classification_report,accuracy_score
def mymodel(model):
    model.fit(X_train,y_train)        #This line trains the model on the training data X_train and y_train
    y_pred = model.predict(X_test)    #This line uses the trained model to predict labels for the test data X_tes
 
    print("Accuracy Score :",accuracy_score(y_test,y_pred))
    print(classification_report(y_test,y_pred))
    
    train = model.score(X_train,y_train)
    test = model.score(X_test,y_test)
    print("Training Performance :",train)
    print("Testing Performance :",test)
    return model
```
